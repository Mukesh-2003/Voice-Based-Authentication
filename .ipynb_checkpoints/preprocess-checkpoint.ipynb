```python
## Mel-Frequency Cepstral Coefficients (MFCCs)

import librosa
import librosa.display
import matplotlib.pyplot as plt

def extract_mfcc(audio_path):
    # Load the audio file
    audio_data, sr = librosa.load(audio_path, sr=16000)
    
    # Extract MFCCs
    mfcc = librosa.feature.mfcc(y=audio_data, sr=sr, n_mfcc=13)
    
    # Plot MFCCs as a heatmap
    plt.figure(figsize=(10, 6))
    librosa.display.specshow(mfcc, x_axis='time', sr=sr)
    plt.colorbar(format='%+2.0f dB')
    plt.title('MFCC')
    plt.savefig('mfcc_output.png')  # Save as image
    plt.show()
    
    return mfcc

# Example usage
mfcc = extract_mfcc("test\kaviya2.wav")
```

```python
## Log-Mel Spectrograms

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

def extract_log_mel_spectrogram(audio_path):
    # Load the audio file
    audio_data, sr = librosa.load(audio_path, sr=16000)
    
    # Compute the Mel spectrogram
    mel_spectrogram = librosa.feature.melspectrogram(y=audio_data, sr=sr, n_mels=128)
    
    # Convert the Mel spectrogram to decibels (logarithmic scale)
    log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)
    
    # Plot the log-Mel spectrogram
    plt.figure(figsize=(10, 6))
    librosa.display.specshow(log_mel_spectrogram, x_axis='time', y_axis='mel', sr=sr)
    plt.colorbar(format='%+2.0f dB')
    plt.title('Log-Mel Spectrogram')
    plt.savefig('log_mel_spectrogram.png')  # Save as image
    plt.show()

    return log_mel_spectrogram

# Example usage
log_mel = extract_log_mel_spectrogram("test\kaviya2.wav")
```

```python
## Pitch and Energy

import librosa
import matplotlib.pyplot as plt

def extract_pitch_and_energy(audio_path):
    # Load the audio file
    audio_data, sr = librosa.load(audio_path, sr=16000)
    
    # Extract pitch (fundamental frequency) using librosa
    pitches, magnitudes = librosa.core.piptrack(y=audio_data, sr=sr)
    pitch = np.mean(pitches, axis=1)
    
    # Extract energy (root mean square)
    energy = librosa.feature.rms(y=audio_data)
    
    # Plot Pitch and Energy
    plt.figure(figsize=(12, 6))
    
    plt.subplot(2, 1, 1)
    plt.plot(pitch)
    plt.title('Pitch')
    
    plt.subplot(2, 1, 2)
    plt.plot(energy[0])
    plt.title('Energy')
    
    plt.tight_layout()
    plt.savefig('pitch_energy_output.png')  # Save as image
    plt.show()

    return pitch, energy

# Example usage
pitch, energy = extract_pitch_and_energy("test\kaviya2.wav")
```

```python
## Voice Activity Detection (VAD) Features

import librosa
import matplotlib.pyplot as plt
import numpy as np

def extract_vad(audio_path):
    # Load the audio file
    audio_data, sr = librosa.load(audio_path, sr=16000)
    
    # Perform Voice Activity Detection (VAD) by using energy thresholding
    energy = librosa.feature.rms(y=audio_data)
    vad = energy > np.mean(energy)  # Simple thresholding method
    
    # Plot the VAD decision
    plt.figure(figsize=(10, 6))
    plt.plot(vad[0], label='Speech (1) vs Silence (0)', color='blue')
    plt.title('Voice Activity Detection (VAD)')
    plt.savefig('vad_output.png')  # Save as image
    plt.show()

    return vad

# Example usage
vad = extract_vad("test\kaviya2.wav")
```

```python
## Delta and Delta-Delta Features

import librosa
import librosa.display
import matplotlib.pyplot as plt

def extract_delta_features(audio_path):
    # Load the audio file
    audio_data, sr = librosa.load(audio_path, sr=16000)
    
    # Extract MFCCs
    mfcc = librosa.feature.mfcc(y=audio_data, sr=sr, n_mfcc=13)
    
    # Compute Delta and Delta-Delta of MFCCs
    mfcc_delta = librosa.feature.delta(mfcc)
    mfcc_delta2 = librosa.feature.delta(mfcc, order=2)
    
    # Plot Delta and Delta-Delta features
    plt.figure(figsize=(10, 6))
    
    plt.subplot(2, 1, 1)
    librosa.display.specshow(mfcc_delta, x_axis='time', sr=sr)
    plt.title('MFCC Delta')
    
    plt.subplot(2, 1, 2)
    librosa.display.specshow(mfcc_delta2, x_axis='time', sr=sr)
    plt.title('MFCC Delta-Delta')
    
    plt.tight_layout()
    plt.savefig('delta_features_output.png')  # Save as image
    plt.show()

    return mfcc_delta, mfcc_delta2

# Example usage
mfcc_delta, mfcc_delta2 = extract_delta_features("test\kaviya2.wav")
```
